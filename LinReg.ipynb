```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
```

```python
# 1 feature, 100 samples, with some noise
X, y = make_regression(n_samples=100, n_features=1, noise=15, random_state=42)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X.shape, y.shape
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

linreg = LinearRegression()
linreg.fit(X_train, y_train)

# Predictions
y_pred = linreg.predict(X_test)
```

```python
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("R² Score:", r2_score(y_test, y_pred))
```

```python
plt.scatter(X.ravel(), y, color='blue', label='Data Points')
plt.plot(X, linreg.predict(X), color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```

```python
# Ridge Regression (L2)
ridge = Ridge(alpha=10)
ridge.fit(X_train, y_train)
ridge_pred = ridge.predict(X_test)

# Lasso Regression (L1)
lasso = Lasso(alpha=10)
lasso.fit(X_train, y_train)
lasso_pred = lasso.predict(X_test)

print("Ridge R² Score:", r2_score(y_test, ridge_pred))
print("Lasso R² Score:", r2_score(y_test, lasso_pred))
```

```python
plt.scatter(X, y, color='lightgray', label='Data')
plt.plot(X, linreg.predict(X), color='red', label='Linear')
plt.plot(X, ridge.predict(X), color='green', linestyle='--', label='Ridge (alpha=10)')
plt.plot(X, lasso.predict(X), color='blue', linestyle='--', label='Lasso (alpha=10)')
plt.title('Comparison of Linear, Ridge, and Lasso Regression')
plt.legend()
plt.show()
```
